```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Linear Regression

```python
num_of_points = 300
# y = a*x + b
a = 0.38
b = 0.56
```

```python
x_data = np.random.normal(0.0, 0.5, num_of_points)
y_data = a * x_data + b + np.random.normal(0.0, 0.1, num_of_points)
```

```python
plt.plot(x_data, y_data, 'o', label='Input Data')
plt.legend()
plt.show()
```

```python
A = tf.Variable(tf.random_uniform([1], minval=-1.0, maxval=1.0))
B = tf.Variable(tf.zeros([1]))
y = A * x_data + B
loss = tf.reduce_mean(tf.square(y - y_data))  # mse
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(0, 21):
        sess.run(train)
        if step % 5 == 0:
            plt.plot(x_data, y_data, 'o', label='step = {}'.format(step))
            plt.plot(x_data, sess.run(A)*x_data+sess.run(B))
            plt.legend()
            plt.show()
    print('a =', sess.run(A))
    print('b =', sess.run(B))
```
